```python
import requests
from bs4 import BeautifulSoup
from soynlp.noun import LRNounExtractor_v2
from collections import Counter

def fetch_and_parse_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        outer_div = soup.find('div', class_='ul_respon tbl_num04')
        
        if outer_div:
            titles = []  
            for ul in outer_div.find_all('ul'):
                a_tag = ul.find('a', class_='detailLinkNologin')
                if a_tag:
                    title = a_tag.text.strip()
                    titles.append(title) 
            return titles 
        else:
            print('"ul_respon tbl_num04" 클래스를 가진 div를 찾을 수 없습니다.')
            return None
    else:
        print("웹 페이지를 로드하는 데 실패했습니다.")
        return None

def main():
    base_url = 'https://web.kangnam.ac.kr/menu/c3052250191ee980c34d207bbedda387.do'
    pages_to_fetch = 20
    all_titles = []

    for page in range(1, pages_to_fetch + 1):
        url = f"{base_url}?paginationInfo.currentPageNo={page}&searchMenuSeq=0&searchType=ttl&searchValue="
        titles = fetch_and_parse_page(url)
        
        if titles:
            all_titles.extend(titles) 

    # soynlp를 사용하여 추출
    noun_extractor = LRNounExtractor_v2(verbose=False)
    nouns = noun_extractor.train_extract(all_titles)  # 명사 추출
    
    # 빈도수 계산
    noun_counts = Counter(nouns)
    most_common_nouns = noun_counts.most_common(30)
    
    print("가장 빈도가 높은 명사 30개:")
    for noun, freq in most_common_nouns:
        print(f"{noun}: {freq}")

if __name__ == "__main__":
    main()
```
